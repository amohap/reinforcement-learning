```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
epsilons=["0.10000", "0.20000", "0.90000"]
betas=["0.00005", "0.00050", "0.00500"]
gammas=["0.80000", "0.85000", "1.00000"]
etas=["0.00350", "0.00500", "0.00700"]
```

# 1st stage

```python
for epsilon in epsilons:
    for eta in etas:
        delta = pd.read_csv('ep%s_be0.00005_ga0.85000_et%s/Delta_save_SARSA.csv' %(epsilon,eta))
        delta['Delta_save'] = delta['Delta_save'].ewm(span=2000, adjust=False).mean()
        plt.plot(delta['Delta_save'], label='ep%s_be0.00005_ga0.85000_et%s' %(epsilon,eta))

plt.xlabel('Episodes')
plt.ylabel('Error')
plt.title('Average Loss')
plt.legend()
# plt.show()
plt.savefig('report_delta.png')
plt.show()
plt.close()
        
    
        
```

```python
for epsilon in epsilons:
    for eta in etas:
                r = pd.read_csv('ep%s_be0.00005_ga0.85000_et%s/R_save_SARSA.csv' %(epsilon,eta))
                r['Delta_save'] = r['R_save'].ewm(span=2000, adjust=False).mean()
                plt.plot(r['R_save'], label='ep%s_be0.00005_ga0.85000_et%s' %(epsilon,eta))

plt.xlabel('Episodes')
plt.ylabel('Reward')
plt.title('Average Rewards per Episode')
plt.legend()
plt.savefig('report_reward.png')
plt.show()
plt.close()
```

```python
for gamma in gammas:
    for beta in betas:
        m = pd.read_csv('ep0.20000_be%s_ga%s_et0.00350/N_moves_SARSA.csv' %(beta,gamma))
        m['N_moves'] = m['N_moves'].ewm(span=2000, adjust=False).mean()
        plt.plot(m['N_moves'], label='be%s_ga%s' %(beta,gamma))

plt.xlabel('Episodes')
plt.ylabel('Number of Steps until "Done"')
plt.title('Average Number of Steps until "Done" per Episode')
plt.legend()
plt.show()
plt.savefig('report_moves.png')
plt.close()
```
